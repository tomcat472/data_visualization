```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# matplotlib တွင် အသုံးပြု၍ ရသော style များအား ကြည့်ခြင်း

```python
# matplotlib မှာ အသုံးပြု၍ ၍သော style များအား ကြည့်ခြင်း
plt.style.available
```

```python
# matplotlib style များအား ကြည့်ခြင်း
plt.style.use('seaborn-v0_8-dark-palette')
```

```python
x=np.linspace(0,10,1000) #linspace နဲ့ 0 ကနေ 10 အတွင်း ဂဏန်းအလုံးရေ 1000 ကိုယူခြင်း
# np.cumsum က cummulative sum ကိုတွက်တာ
# np.random.randn က 0 ကနေ 1 ကြား အလုံး 1000 , အနောက်က 5 က column အရေအတွက်
# 0 က cumsum လုပ်တဲ့အခါ သုံးတဲ့ axis

y=np.cumsum(np.random.randn(1000,5),0)
plt.plot(x,y)
plt.xlabel('np.linspace(0,10,1000)')
plt.ylabel('np.cumsum(np.random.randn(1000,5),0)')
plt.title('Sample plot')

plt.legend(['apple','orange','grape','banana','melon'] #legend name ထည့်ခြင်း
           ,ncol=2 #legend name column ထည့်ခြင်း
           ,loc=0 #legend location
           )
```
