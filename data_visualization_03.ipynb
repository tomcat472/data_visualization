```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(10)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
# xlabel / ylabel က axis label သတ်မှတ်တာဖြစ်တယ်။
plt.xlabel('x=np.arange(10)')
plt.xlim([2,7.5])

plt.ylabel('y=f(x)')
plt.ylim([-1,100])

# title က ခေါင်းစဉ်ကို ထည့်တာဖြစ်ပါတယ်။
plt.title('Simple Plot');
```

```python
x=np.arange(10)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)

# legend ထည့်ခြင်း
plt.legend(['x**2','x**3','2*x','2**x'],loc=0)

plt.xlabel('x=np.arange(10)')
plt.xlim([2,7.5])

plt.ylabel('y=f(x)')
plt.ylim([-1,100])

plt.title('Simple Plot');
```

# Colors, Styles and Markers

```python
x=np.arange(5)
y=x
plt.plot(x,y+1,'g')
plt.plot(x,y+0.5,'y')
plt.plot(x,y,'r')
plt.plot(x,y-0.2,'c')
plt.plot(x,y-0.4,'k')
plt.plot(x,y-0.6,'m')
plt.plot(x,y-0.8,'w')
plt.plot(x,y-1,'b');
```

```python
plt.plot(x,y,color='g',linestyle='--',linewidth=.5,marker='*',
         markerfacecolor='r',markeredgecolor='k',markeredgewidth=1,markersize=10)
```
