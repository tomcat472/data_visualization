# Pandas Plotting

```python
import matplotlib.pyplot as plt
plt.close('all')
```

# Basic Pandas Plotting

```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# pandas date range
pd.date_range('12/3/2020',periods=1000)
```

```python
# plotting timeseries data
ts=pd.Series(np.random.randn(1000),pd.date_range('12/3/2020',periods=1000))
ts.cumsum().plot();
```

```python
ts.index
```

```python
df=pd.DataFrame(np.random.randn(1000,5),index=ts.index,columns=list('ABCDE'))
plt.figure()
df.cumsum().plot();
```

# Barplot

```python
df.iloc[5].plot(kind='bar');
```

```python
df=pd.DataFrame(np.random.rand(10,6),columns=list('ABCDEF'))
df.plot.bar()
```

```python
df=pd.DataFrame(np.random.rand(10,6),columns=list('ABCDEF'))
df.plot(kind='bar')
```

```python
df=pd.DataFrame(np.random.rand(10,6),columns=list('ABCDEF'))
df.plot.bar(stacked=True);
```

```python
df=pd.DataFrame(np.random.rand(10,6),columns=list('ABCDEF'))
df.plot.barh(stacked=True);
```

# Histogram

```python
df.A.hist(bins=10);
```

```python
df=pd.DataFrame({
    'a':np.random.randn(1000)+1,
    'b':np.random.randn(1000),
    'c':np.random.randn(1000)-1
},columns=list('abc')) #columns=['a','b','c']

plt.figure()
df.plot.hist(alpha=0.5);
```

```python
plt.figure()
df.plot.hist(stacked=True,bins=30);
```

```python
df.hist(grid=False,color='k',bins=40,alpha=0.5);
```

# Area Plot

```python
df=pd.DataFrame(np.random.rand(1000,5),columns=list('abcde'))
df.plot.area();
```

# Scatter Plot

```python
plt.scatter(df.a,df.b);
```

```python
df.plot.scatter(x='a',y='b');
```

```python
ax=df[750:].plot.scatter(x='a',y='b',color='#7e46f0',label='groupA');
df[750:].plot.scatter(x='c',y='d',color='#e637a3',label='groupB',ax=ax);
```
