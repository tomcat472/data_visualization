```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
# swarm plot
sns.catplot(data=tips,kind='swarm',x='day',y='total_bill',hue='smoker')
```

```python
sns.catplot(data=tips,kind='violin',x='day',y='total_bill',hue='smoker',split=True)
```

```python
sns.catplot(data=tips,kind='bar',x='day',y='total_bill',hue='smoker')
```

# Multivariable views on complex datasets

```python
# jointplot
sns.jointplot(data=tips,x='total_bill',y='tip',hue='smoker')
```

```python
tips
```

```python
# pairplot
sns.pairplot(data=tips,hue='sex')
```

```python
df=sns.load_dataset('penguins')
df.head()
```

```python
sns.histplot(
    data=df,
    x='flipper_length_mm',
    hue='species',
    multiple='stack'
)
```

```python
sns.kdeplot(
    data=df,
    x='flipper_length_mm',
    hue='species',
    multiple='stack'
)
```

# Data Structure accepted by seaborn

```python
df=sns.load_dataset('flights')
df.head()
```

```python
sns.relplot(data=df,x='year',y='passengers',kind='line',hue='month')
```

```python
df_pivot=df.pivot(index='year',columns='month',values='passengers')
df_pivot
```

```python
sns.relplot(data=df_pivot,kind='line')
```
