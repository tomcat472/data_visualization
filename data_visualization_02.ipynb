```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Basic Plot

```python
A=np.arange(10)
print(A)
print(type(A))
```

```python
A.ndim
```

```python
x=A
y=A**2
plt.plot(x,y,marker='o',linestyle='--')
plt.show()
```

```python
# line style တွေ marker style တွေကို သိချင်ရင် plt.plot? လို့ရိုက်ပြီး help မှာဖတ်လို့ရပါတယ်။
```

# Shorthands

```python
# plt.plot(x,y,color='red',marker='o',linestyle='--');
plt.plot(x,y,'ro--');
```

# Exercise 2.1

```python
x=np.linspace(1,10,100)
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
plt.plot(x,a,'k:')
plt.plot(x,y,'b-')
plt.plot(x,z,'r--');
```

```python
# mutiline plots ပြုလုပ်ခြင်း
x=np.linspace(1,10,100)
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
plt.plot(x,a,'k:',x,y,'b-',x,z,'r--');
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
plt.axis('off') # axis ဖျောက်ခြင်း
plt.plot(x,a,'k:',x,y,'b-',x,z,'r--');
```

```python
x=np.linspace(1,10,100)
y=np.sin(x)
z=np.cos(x)
a=np.tan(x)
plt.axis([3,8,-20,40]) #axis များအား သတ်မှတ်ခြင်း
plt.grid(True) # grid line ထည့်ခြင်း
plt.plot(x,a,'k:',x,y,'b-',x,z,'r--');
# plt.savefig('abc.png') #ပုံ save ခြင်း
```
