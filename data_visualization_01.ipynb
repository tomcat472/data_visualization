```python
# importing library
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# seaborn ၏ default dataset အား အသုံးပြုခြင်း
df=sns.load_dataset('diamonds')
```

```python
df.head()
```

# Basic Matplotlib Plot

```python
# matplotlib ၏ line plot အားအသုံးပြုခြင်း
plt.plot(df.price.head(20))
```

```python
# line plot အသုံးပြုရာတွင် x,y ဖြင့်လည်း အသုံးပြုလို့ရပါသည်။
# အရှေ့မှ တန်ဖိုးသည် x ဖြစ်ပြီး အနောက်မှ တန်ဖိုးသည် y ဖြစ်ပါသည်။
plt.plot(df.price.tail(20),df.table.tail(20))
```

```python
# scatter plot
# scatter plot လုပ်ရာတွင် x,yတန်ဖိုးလိုအပ်ပါသည်။
plt.scatter(df.price.head(200),df.table.head(200));
```

# Creating Figure Objects and Axes Subplots

```python
import numpy as np
```

```python
A=np.random.rand(100)
```

```python
plt.plot(A)
```

```python
plt.plot(np.sin(A))
```

```python
plt.plot(A,color='red')
plt.plot(np.sin(A),color='black')
```

```python
# ပထမဆုံး figure object တစ်ခုကို ပြုလုပ်ရမယ်။
fig=plt.figure()

# ပြီးရင် ကိုယ်ထားချင်တဲ့ axes တွေကို သတ်မှတ်မယ်။
axes1=fig.add_subplot(2,1,1) #2,1 က 2 row 1 col, အနောက်က 1 က number of picture ဖြစ်ပါတယ်။
plt.plot(A,color='red')

axes2=fig.add_subplot(2,1,2)
plt.plot(np.sin(A),color='black')

plt.show()
```

```python
# ပထမဆုံး figure object တစ်ခုကို ပြုလုပ်ရမယ်။
fig=plt.figure(figsize=(10,3.5))

# ပြီးရင် ကိုယ်ထားချင်တဲ့ axes တွေကို သတ်မှတ်မယ်။
axes1=fig.add_subplot(1,2,1)
plt.plot(A,color='red')

axes2=fig.add_subplot(1,2,2)
plt.plot(np.sin(A),color='black');
```

```python
# ဒီနည်းကျတော့ subplot ကို တခါတည်းလုပ်လို့ရပါတယ်။
plt.figure(figsize=(10,3.5))

plt.subplot(1,2,1)
plt.plot(A,color='red')


plt.subplot(1,2,2)
plt.plot(np.sin(A),color='black');
```
